```python
import pandas as pd
import joblib
from sklearn.ensemble import IsolationForest
import os

def train_anomaly_detector_v2():
    print("üß† Entrenando Isolation Forest v2 (Con conocimiento de Leucemia/Anemias)...")
    
    # --- RUTA CORREGIDA DE DATOS ---
    input_path = "../data/nhanes_augmented_data.csv"
    try:
        df = pd.read_csv(input_path)
    except FileNotFoundError:
        print(f"‚ùå Falta '{input_path}'")
        return
    
    X = df.drop(columns=['Diagnosis', 'ID'])
    
    model_if = IsolationForest(contamination=0.05, random_state=42)
    model_if.fit(X)
    
    # --- RUTA CORREGIDA PARA GUARDAR MODELO ---
    if not os.path.exists("../models"):
        os.makedirs("../models")
        
    joblib.dump(model_if, '../models/anomaly_detector_v2.joblib')
    print("‚úÖ Guardado: '../models/anomaly_detector_v2.joblib'")

if __name__ == "__main__":
    train_anomaly_detector_v2()
```
