```python
import pandas as pd
import joblib
from sklearn.ensemble import IsolationForest

def train_anomaly_detector_v2():
    print("üß† Entrenando Isolation Forest v2 (Con conocimiento de Leucemia/Anemias)...")
    
    # Usamos el dataset AUMENTADO
    try:
        df = pd.read_csv("nhanes_augmented_data.csv")
    except FileNotFoundError:
        print("‚ùå Falta 'nhanes_augmented_data.csv'")
        return
    
    X = df.drop(columns=['Diagnosis', 'ID'])
    
    # Aumentamos un poco la contaminaci√≥n porque ahora tenemos patolog√≠as extremas conocidas
    model_if = IsolationForest(contamination=0.05, random_state=42)
    model_if.fit(X)
    
    joblib.dump(model_if, 'anomaly_detector_v2.joblib')
    print("‚úÖ Guardado: 'anomaly_detector_v2.joblib'")

if __name__ == "__main__":
    train_anomaly_detector_v2()
```
